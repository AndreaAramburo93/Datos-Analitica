```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear el DataFrame con los datos
data = {
    "Año": [1993, 1994, 1995],
    "Hawaiian Club": [450000, 475000, 390000],
    "French Riviera": [225000, 240000, 205000],
    "Bahamas Beach": [245000, 255000, 345000]
}

df = pd.DataFrame(data)
df.set_index("Año", inplace=True)

# Graficar líneas
plt.figure(figsize=(8, 5))
ax = df.plot(kind='line', marker='o', legend=True, ax=plt.gca())

# Fijar los ticks del eje X en los años exactos
ax.set_xticks([1993, 1994, 1995])
ax.set_xticklabels([1993, 1994, 1995])

plt.xlabel('Año')
plt.ylabel('Valor')
plt.title('Evolución de Valores por Destino (1993-1995)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Datos
data = {
    "Año": [1993, 1994, 1995],
    "Hawaiian Club": [450000, 475000, 390000],
    "French Riviera": [225000, 240000, 205000],
    "Bahamas Beach": [245000, 255000, 345000]
}
df = pd.DataFrame(data).set_index("Año")

# Configurar figura y ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Paleta de colores para cada serie
colores = {
    "Hawaiian Club": '#1f77b4',
    "French Riviera": '#ff7f0e',
    "Bahamas Beach":  '#2ca02c'
}

# Graficar cada serie con marcadores y grosor adecuado
for destino, color in colores.items():
    ax.plot(
        df.index,
        df[destino],
        label=destino,
        color=color,
        linewidth=2,
        marker='o',
        markersize=6,
        markerfacecolor='white',
        markeredgecolor=color,
        markeredgewidth=1.5
    )
    # Anotar el valor encima de cada punto
    for año, valor in zip(df.index, df[destino]):
        ax.text(
            año,
            valor + df[destino].max() * 0.02,
            f'${valor:,.0f}',
            ha='center',
            va='bottom',
            fontsize=8,
            color=color
        )

# Formato de miles en eje Y
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))

# Rejilla suave (solo horizontal)
ax.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.3)
ax.grid(axis='x', visible=False)

# Ticks del eje X en los años
ax.set_xticks([1993, 1994, 1995])
ax.set_xticklabels([1993, 1994, 1995], fontsize=10)

# Leyenda fuera del área de dibujo
ax.legend(
    title='Destino',
    loc='upper center',
    bbox_to_anchor=(0.5, 1.12),
    ncol=3,
    fontsize=10,
    title_fontsize=11
)

# -------------------------------------------------------
# Aquí ajustamos espaciados para el título y los ejes:
# -------------------------------------------------------

# 1) Título más arriba (pad=20 aumenta la separación del título respecto al gráfico)
ax.set_title(
    "Evolución de Valores por Destino (1993–1995)",
    fontsize=18,
    pad=40
)

# 2) Etiquetas de ejes más abajo (labelpad=15 separa la etiqueta del eje de los ticks/valores)
ax.set_xlabel(
    "Año",
    fontsize=11,
    labelpad=30,
)
ax.set_ylabel("Valor (USD)", fontsize=11, labelpad=15)

# -------------------------------------------------------

plt.tight_layout()
plt.show()
```
